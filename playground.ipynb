```python
# %%
import numpy as np
%matplotlib inline
# %matplotlib widget
import matplotlib.pyplot as plt
```

```python
Nmax= 1
Nmin= 0.01
istep = 100
decay_rate = 0.1
```

```python
ite = np.arange(istep)
print('ite: ', ite)
eps_exp = (Nmax-Nmin)*np.exp(-decay_rate*ite)
eps_lin = np.linspace(Nmax, Nmin, istep)
```

```python


plt.figure(figsize=(5, 3))
plt.plot(ite, eps_exp, label=f'Exponential (decay={decay_rate})', linewidth=2)
plt.plot(ite, eps_lin, label='Linear Decay', linestyle='--')

plt.title('Epsilon Decay: Exponential vs. Linear')
plt.xlabel('Episode')
plt.ylabel('Epsilon (Probability of Exploration)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
